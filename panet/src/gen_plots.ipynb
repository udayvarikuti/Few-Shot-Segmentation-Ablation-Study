```python
import json
import numpy as np
import matplotlib.pyplot as plt 
import os
import shutil
```

```python
export_dir="./Plots/"
if(os.path.isdir(export_dir)):
    shutil.rmtree(export_dir)
os.mkdir(export_dir)
```

```python
cfg_ids=["0806141209",
"0906012133",
"0906030148",
"0906061439",
"0906233445"]

for ids in cfg_ids:
    dir= "../test_models/info/cfg_"+ids+".json"
    openfile=open(dir, "r")
    config= json.load(openfile)
    LossSeg=config["losses"]["loss_query"]
    LossPAR=config["losses"]["loss_PAR"]
    plt.figure(ids)
    plt.plot(LossSeg)
    plt.plot(LossPAR)
    plt.title(config["desctiption"])
    plt.xlabel("Iterations (x100) ->")
    plt.ylabel("Loss ->")
    plt.legend(["Segmentation Loss","Alignment Loss"])
    plt.savefig(export_dir+"_"+ids+"_.jpg")
```
