```python
import torch
import torch.nn as nn
```

```python
model = torch.hub.load('pytorch/vision:v0.10.0', 'deeplabv3_resnet101', pretrained=True)
```

```python
class Identity(nn.Module):
    def __init__(self):
        super(Identity, self).__init__()

    def forward(self, x):
        return x
```

```python
model.aux_classifier=Identity()
print(model)
```
